## Task 1. Linear Ridge Regression (1 point)
Let us consider the problem of linear ridge regression for data $(x_{1},y_{1}),\dots,(x_{n},y_{n})\in\mathbb{R}^{d\times 1}$. Let the objects have positive **sample weights** $v_{i}>0$, i.e. the optimization problem is
$$\sum_{i=1}^{n}v_{i}\cdot L(y_{i}, \hat{y}_{i})+\frac{\lambda}{2}\|w\|_{2}^{2}=\sum_{i=1}^{n}v_{i}\cdot (\boldsymbol{w}\cdot\boldsymbol{x}_{i}-y_{i})^{2}+\frac{\lambda}{2}\|w\|_{2}^{2}\rightarrow \min_{\boldsymbol{w}}.$$
This problem reduces to classical linear ridge regression when $v_{i}\equiv 1$. The matrix form of the problem is
$$(Xw^{\top}-y)^{\top}V(Xw^{\top}-y)+\frac{\lambda}{2}w^{\top}w\rightarrow\min_{\beta},$$
where $V=V^{\top}\in\mathbb{R}^{n\times n}$ is the diagonal matrix with diagonal elements $v_{1},\dots, v_{n}$. Note that the quadratic problem is still convex (w.r.t. $\boldsymbol{w}$), thus, the solution is unique. Solve this problem for any symetric positive-defined matrix $V$ (not just diagonal) and provide the answer in the matrix form.
### Your solution:

$$
(Xw - y)^T V (Xw - y) = w^T V X^T w - y^T V X w - w^T X^T V y + y^T V y
    = y^T V y - 2 w^T X^T V y + w^T V X^T w
$$
$$
   y^T V y - 2 w^T X^T V y + w^T V X^T w + \frac{\lambda}{2}w^T w \rightarrow \min_{w}
$$
$$
\triangledown_{w} (y^T V y - 2 w^T X^T V y + w^T V X^T w + \frac{\lambda}{2}w^T w) = 0
    \Rightarrow 2 X^T V X w - 2 X^T V y + \lambda w = 0
$$

Result:
$$
\big [X^T V X w + \frac{\lambda}{2} I \big] \cdot w = X^T V y
    \Rightarrow w = \big [X^T V X w + \frac{\lambda}{2} I \big]^{-1} X^T V y
$$

## Task 2. Logistic Regression (1 point)
Let us consider the case when in the problem of binary classification the training set is lineary separable. Show that in this case the optimization problem for logistic regression **without L2-regularization** does not have optimum.
### Your solution:

Let's consider logistic regression, denote sigmoid function as $g(z)$. 
Since, $1 - g(z) = g(-z)$ we get $1 - g(\theta^T \boldsymbol{x} + \theta_0) = g(-(\theta^T \boldsymbol{x} + \theta_0))$.

So, let's $y \in \{-1, 1\}$ $\rightarrow \mathbf{P}(y | \boldsymbol{x}, \theta, \theta_0) = g(y(\theta^T \boldsymbol{x} + \theta_0))$

To get optimal params we should maximize likelihood. Drop some steps and write the following formula:
$$
\sum_{t=1}^n \log[1 + \exp(-y_t(\theta^T \mathbf{x}_t + \theta_0))] \rightarrow \min_{\theta}
$$

If training examples are linearly separable, we can find parameter values such that $y_t(\theta^T x_t + \theta_0)$ are positive for all training examples. By scaling up the parameters, we make these values larger and larger. So $\log[1 + \exp(-y_t(\theta^T \mathbf{x}_t + \theta_0))]$ is a strictly decreasing as a function of $y_t(\theta^T x_t + \theta_0)$.

It means infinite scaling of any parameters would attain the highest likelihood. So, the maximum likelihood parameter values would become unbounded.

## Task 3. Bayesian Naive Classifier (1 point)
Show that in case of $d$ binary-valued features $x_{j}\in\{0, 1\}$ (for $j=1,2,\dots,d$) Bayesian Naive Classifier's decision rule is linear.
### Your solution:

$$
\mathbf{P}(y = 1 | \boldsymbol{x})
    = \frac{\mathbf{P}(\boldsymbol{x}, y = 1)}{\mathbf{P}(\boldsymbol{x}, y = 1) + \mathbf{P}(\boldsymbol{x}, y = -1)}
    = \frac{1}{1 + \exp(-f(\boldsymbol{x}))}
$$

where $f(\boldsymbol{x}) = \log \frac{\mathbf{P}(\boldsymbol{x}, y = 1)}{\mathbf{P}(\boldsymbol{x}, y = -1)}$

If $f(x)$ is a linear function, then the form of the expression above would be exactly as logistic regression.

$$
f(\boldsymbol{x}) = \sum_{i \in \mathcal{J}}\log \frac{\mathbf{P}(x_i, y = 1)}{\mathbf{P}(x_i, y = -1)}
    + \frac{\mathbf{P}(y = 1)}{\mathbf{P}(y = -1)}
$$

Now, assume that features are binary:
$$
\log \frac{\mathbf{P}(x_i, y = 1)}{\mathbf{P}(x_i, y = -1)}
    = x_i \log \frac{\mathbf{P}(x_i = 1, y = 1)}{\mathbf{P}(x_i = 1, y = -1)}
    + (1 - x_i) \log \frac{\mathbf{P}(x_i = 0, y = 1)}{\mathbf{P}(x_i = 0, y = -1)}
$$

So,
$$
f(\boldsymbol{x}) = \hat{\theta} \boldsymbol{x} + \hat{\theta}_0
$$

## Task 5. Bootstrap (1 point)
Let the subsample $\hat{X}^{n}$ of size $n$ be generated from sample $X^{n}=\{\boldsymbol{x}_{1},\dots\boldsymbol{x}_{n}\}$ by bootstrap procedure. Find the probability of object $x_{i}$ to be present in $\hat{X}^{n}$ and compute the limit of this probability for $n\rightarrow\infty$.
### Your solution:

So, probability that $\boldsymbol{x}_i$ isn't included in bootstrap $\hat{X}^n$ is equal to: 
$$
\mathbf{P}(\boldsymbol{x}_i \not \in \hat{X}^n) 
    = \big (\frac{n - 1}{n} \big )^n
    = \big (1 - \frac{1}{n} \big )^n
    \to \frac{1}{e}, \text{ if } n \to \infty
$$

Probability that $\boldsymbol{x}_i$ is included in bootstrap $\hat{X}^n$ is equal to: 
$$
\mathbf{P}(\boldsymbol{x}_i \in \hat{X}^n)
    = 1 - \mathbf{P}(\boldsymbol{x}_i \not \in \hat{X}^n)
    = 1 - \big (\frac{n - 1}{n} \big )^n
    = 1 - \big (1 - \frac{1}{n} \big )^n
    \to 1 - \frac{1}{e}, \text{ if } n \to \infty
$$

$$\hat{y}=\frac{1}{n}\sum_{i=1}^{n}y_i.$$

In this case, for simplicity you may assume that there is only one most frequent label.

## Task 9. Kernel Methods (1 point)
Prove that Gaussian Kernel $K(x,x')=\exp(-\|x-x'\|^{2})$ is positive definite.
### Your solution:

Represent this kernel $K(x,x')=\exp(-\|x-x'\|^{2})$ in another form:
$$
K(x,x')
    =\exp(-\|x-x'\|^{2})
    =\exp(-\frac{1}{2}x^T x + x^T x' - \frac{1}{2} {x'}^T x')
$$
Denote $f(x) = \exp(-\frac{1}{2}x^T x)$:
$$
K(x,x')
    = f(x) \cdot \exp(x^T x') \cdot f(x')
$$

Let's proof that $K_{exp}(x, x') = \exp(x^T x')$ is positive-semidefinite:
$$
K_{exp}(x, x')
    = \exp(x^T x')
    = \sum_{k=0}^{\infty} \frac{(x^T x')^k}{k!}
$$
$\mathbf{G}$ (Gramm) is a matrix coresponding to $x^T x'$ kernel. $\mathbf{G}$ is positive-semidefinite. It means that matrix corresponding to exponential kernel $\mathbf{K}_{exp}$ is also positive-semidefinite

Consider $K(x,x')$ and it's matrix $\mathbf{K}$:
$$
\forall c \in \mathbf{R}^{n \times 1}, 
c^T \mathbf{K} c 
    = \sum_{i,j = 1}^n c_{i} c_{j} K(x_i, x_j)
    = \sum_{i,j = 1}^n (c_{i}f(x_i)) (c_{j}f(x_j)) K_exp(x_i, x_j)
$$
Since $K_exp(x, x')$ is positive-semidefinite:
$$
\forall \hat {c} \in \mathbf{R}^{n \times 1}, \sum_{i,j = 1}^n \hat{c_{i}} \hat{c_{j}} K_{exp}(x_i, x_j) \geq 0
$$
So, $K(x, x')$ is positive-semidefinite
