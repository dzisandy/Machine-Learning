## Task 1. Numpy Problem 1 (1 point)

Write a function, which takes a matrix `arr` and centers each of its rows
by the mean within that row. Check [this](https://docs.scipy.org/doc/numpy/reference/routines.statistics.html)
out for documentation.

```python
def center(arr):
    """Center each row of a matrix by the mean value in that row.

    Parameters
    ----------
    arr : array, shape = (n_rows, n_cols)
        The matrix, the rows of which to center.

    Returns
    ----------
    out : array, shape = (n_rows, n_cols)
        The final row-centered matrix.
    """
    assert arr.ndim == 2
    n_rows, n_cols = arr.shape

    ### BEGIN Solution
    rows_mean = np.mean(arr, axis=1).reshape((n_rows, 1))
    out = arr - rows_mean
    ### END Solution

    return out
```

```python
arr = np.array([[1, 2, 3, 4], [5, 6, 7, 10]])
print('Initial array: ')
print(arr)
print('Centered array: ')
print(center(arr))
```

## Task 2. Numpy Problem 2 (1 point)

Implement the following fancy function:
\begin{equation}
    f(x) = \sigma\bigl(
                \max\{x + 5, 0\} + \max\{5 - x, 0\}
                + \max\{\min\{\cos(2 x \pi), \tfrac12\}, -\tfrac14\}
            \bigr)
         \,,
\end{equation}
where $\sigma(x) = (1+e^{-x})^{-1}$ is the sigmoid function.

```python
def fancy_function(x):
    """Compute some fancy function.
    
    Parameters
    ----------
    x : array, 1 dimendional, shape=(n_samples,)
        The array argument values.

    Returns
    -------
    y : array, 1 dimendional, shape=(n_samples,)
        The values of the fancy function.
    """
    
    ### BEGIN Solution
    def sigmoid(var):
        return 1./(1 + np.exp(-var))
    
    comp_var = np.maximum(x+5, 0) + np.maximum(5-x, 0) + np.maximum(np.minimum(np.cos(2*x*np.pi), 1./2), -1./4)
    out = sigmoid(comp_var)
    ### END Solution
    
    return out
```

Plot the function

```python
x = np.linspace(-12.5, 12.5, num=1001)
fig, ax = plt.subplots(figsize=(5, 5))
ax.plot(x, fancy_function(x))
plt.show()
```

### Subproblem 4.2. Pipeline: Fitting to Data by Feature Engeneering (1 of 2 points)
In previous task 1.1 several classifiers obviously failed fitting to data. This happend because the decision rule of the classifier has a restricted form (e.g. linear for linear models), while the data is more complicated.

One may try to change the parameters of the classifier (e.g. increase the number of trees in Forest) in order to improve accuracy, but some models (especially linear) do not have parameters that can change the form of the decision rule.

In this case, the **feature engeneering** helps: one may try to compute new (e.g. non-linear) features based on the existing pool and fit the classifier in the new features. This may help low-complex claffifiers to fit to hard data dependencies.

Your task it to
* Choose **two classifiers** from the following: Decision Tree, Random Forest, Naive Bayes, Logistic Regression;
* By generating of additional features (e.g. polynomial) make them achieve accuracy $>0.95$.
* For each classifier, write 2-3 sentences why did you decide to choose these features.
* Plot their decision rules in the original feature space.

It is your choice how to generate features. You may create hand-crafted featues and add them manualy. Nevertheless, we **highly suggest** to get used to and apply the following builtin `sklearn` methods:
* `PolynomialFeatures`, `GaussianRandomProjection` among others - for feature generation
* `StandartScaler`, `MinMaxScaler` among others - for feature scaling
* `Pipeline` - for combining several operations in a row (e.g. feature creation & prediction)

#### Solution 4.2.

The distribution of samples is obviously nonlinear. So one has to create some additional features. Moreover according to the distribution of samples (which can be seen on the plot) one can assume that the line defined by polinom might separate classes well.

For Logistic Regression polynomial features will provide ability to catch the 'quasi-polinomial' nature of samples and build more precise (than the initial ordinary straight line) separating hyperplane.

For Random Forest polynomial features will help to discover better the complex structure of dataset (especially in the place where samples from two classes intersect at the center on the plot) and build more precise separating hyperplane.

```python
degrees = np.arange(1, 7)
clfs = [
    RandomForestClassifier(n_estimators=25, max_depth=4, n_jobs=-1, random_state=RANDOM_STATE),
    LogisticRegression(random_state=RANDOM_STATE, n_jobs=-1)
]

fig, axes = plt.subplots(
    ncols=2,
    nrows=len(degrees), figsize=(12, 5 * (len(degrees) + 1)),
    dpi=75
)

for degree_num, degree in enumerate(degrees):
    for clf_num, clf in enumerate(clfs):
        pipeline = Pipeline([('features', PolynomialFeatures(degree)),
                        ('scaler', StandardScaler()), 
                        ('clf', clf)])
        pipeline.fit(X, y)
        plot_decision_regions(X=X, y=y, clf=pipeline, legend=2, ax=axes[degree_num, clf_num])
        axes[degree_num, clf_num].set_title(type(clf).__name__ + 
                                       ', accuracy = {}'.format(accuracy_score(y_pred=pipeline.predict(X), y_true=y)) + 
                                        ', degree = {}'.format(degree))
fig.tight_layout()
```

As it can be seen from the plots and scores above the smallest degree for polynomial features for Random Forest which helps to achieve accuracy = 0.952 > 0.95 equals 2. For Logistic Regression the smallest degree is 5 which helps to achieve accuracy = 0.952 > 0.95.

So the final solution to this subproblem is the following:

```python
### BEGIN Solution
clfs = [
    RandomForestClassifier(n_estimators=25, max_depth=4, n_jobs=-1, random_state=RANDOM_STATE),
    LogisticRegression(random_state=RANDOM_STATE, n_jobs=-1)
]

fig, axes = plt.subplots(
    ncols=2,
    nrows=1, figsize=(12, 6),
    dpi=75
)

for num, clf, degree in zip(range(2), clfs, [2, 5]):
    pipeline = Pipeline([('features', PolynomialFeatures(degree)),
                        ('scaler', StandardScaler()), 
                        ('clf', clf)])
    pipeline.fit(X, y)
    plot_decision_regions(X=X, y=y, clf=pipeline, legend=2, ax=axes[num])
    axes[num].set_title(type(clf).__name__ + 
                                       ', accuracy = {}'.format(accuracy_score(y_pred=pipeline.predict(X), y_true=y)) + 
                                        ', degree = {}'.format(degree))
fig.tight_layout()
### END Solution
```

With such chosen features both models show acceptable accuracy and don't suffer from overfitting

## Task 5. Model Selection (1+1 points)

You are to test Random Forests and Support Vector Machines on a trivial [Tic Tac Toe Endgame Dataset](https://datahub.io/machine-learning/tic-tac-toe-endgame). Let's load it.

```python
data = pd.read_csv('data/tic-tac-toe.csv')
X, y = data.drop('class', axis=1), data['class'].astype(int)
data.sample(3)
```

The dataset consists of several possible endgame positions of the Tic-Tac-Toe game. The target variable is the victory of `x` player over `o` player (victory or defeat/draw). Since the features are categorical, we simply transform them to real-valued $-1$ for `o`, $1$ for `x` and $0$ for emply cell `b`.

```python
X = X.applymap(lambda v: 1 if v == 'x' else -1 if v == 'o' else 0)
X.sample(4).sort_index()
```

We are going to test how machine learning algorithms can classify the final game positions into the ones when `x` player won and all others. Everybody knows that for this problem there is a simple decision rule: `x` wins if there are three `x`'s in a row/column/diagonal. But can our cool machine learning tools catch this trivial dependence? In this problem the class balance is around $2:1$ so we still use accuracy metric.

Intuitively, this rule is logical, i.e. one may expect decision-tree-based algorithm to be the most appropriate for this case. But is that true? In the code below we compare huge Random forest with simple Logistic Regression and SVM with default parameters.

```python
rf = RandomForestClassifier(n_estimators=400, max_depth=10)
svm = SVC(gamma='auto')
lr = LogisticRegression(solver='lbfgs')
clfs = (rf, svm, lr)

for clf in clfs:
    score = cross_val_score(clf, X, y, cv=5).mean()
    name = clf.__class__.__name__
    print(f'{name} scored {round(score, 3)}')
```

We **counterintuitively** see that RandomForest is outperformed by SVM. Moreover, the simplest Logistic regression significantly outperforms everything. Let's improve Random Forest and SVM in order to be able to compete with Logistic model.

#### Solution 5.2
I decided to make the following features:

## Task 6. Bagging Ensembles of Regressors (2 points)
In this problem you are to deal with [Concrete Compressive Strength Dataset](https://www.kaggle.com/elikplim/concrete-compressive-strength-data-set). You goal will be to determine the optimal parameters for two Bagging-Based Forest Ensemble **Regressors** and compare the forests. Lets load the data and split it into test and train parts.

```python
data = pd.read_csv('data/concrete.csv').astype(float)
X = data.drop('concrete_compressive_strength', axis=1)
y = data.concrete_compressive_strength

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

data.sample(3).sort_index()
```

Please note that both in Task 4 and Task 5 the whole data was the Train Data. In Task 4 the output score (accuracy) was the **train** score (i.e. the score on the train data of the model fitted on the same data). In task 5 the output score was the **validation score**, i.e. the result of validating the model on the train data.

In this problem, we do a step further and split the whole data into train part (on which we train & validate) and test part (where we compute the final **test** score on the validated model).

In this problem you are to consider the `RandomForestRegressor` and `ExtraTreesRegressor` models for the prediction of concrete compressive strength under **squared loss function** (mean squared error). Recall that Random Forest was discussed in the lectures. [Extremely Randomized Forest](http://citeseerx.ist.psu.edu/viewdoc/download?doi=10.1.1.65.7485&rep=rep1&type=pdf) is an other bootstraped forest with simple tree building algorithm. Basically, each split of each tree node is chosen at random both w.r.t. feature and threshold (while in random forest the split minimizes impurity).

You have to do the following steps and answer the following questions:
* For both Forests perform the Grid Search (on the train data) over most important algorithm's parameters (what are they?) to determine the optimal hyperparameters.
* For the optimal parameters output the train, validation score and the score for predicting for the test data.
* Compare the obtained scores. Explain, why the scores differ a lot for train and validation/test.
* Which of the algorithms perform better on the training set? Explain why!

```python
### BEGIN Solution

# Printing the header of result table
rf_name, rf_name_len = RandomForestRegressor.__name__, len(RandomForestRegressor.__name__)
et_name, et_name_len = ExtraTreesRegressor.__name__, len(ExtraTreesRegressor.__name__)
header, header_len = 'Ensemble Classifier', len('Ensemble Classifier')
header_tail = '| Train | ' + 'Valid | ' + 'Test | ' + 'n_estimators | ' + 'max_depth'
header_tail_len = len(header_tail)
longest_name = np.max([rf_name_len, et_name_len, header_len])
print(header + ' '*(longest_name - header_len + 1) + header_tail)
print('-'*(longest_name + header_tail_len + 2))

# Define regressors and the set of parameters for Grid Search
regressors = [
    ExtraTreesRegressor(random_state=RANDOM_STATE, n_jobs=-1),
    RandomForestRegressor(random_state=RANDOM_STATE, n_jobs=-1)
]
parameters = {
    'n_estimators': np.arange(100, 1100, step=100),
    'max_depth': [3, 5, 7, 9, 11]
}
best_regressors = []

# Perform Grid Search with CV
for regressor in regressors:
    gs = GridSearchCV(regressor, parameters, scoring='neg_mean_squared_error', cv=5, n_jobs=-1)
    gs.fit(X_train, y_train)
    train_score = mse(gs.best_estimator_.predict(X_train), y_train)
    valid_score = -gs.best_score_
    test_score = mse(gs.best_estimator_.predict(X_test), y_test)
    print(regressor.__class__.__name__ + ' '*(longest_name - len(regressor.__class__.__name__) + 1) + '| '  
          + str(round(train_score, 2)) 
          + ' '*(1 + len('Train') - len(str(round(train_score, 2)))) + '| '
          + str(round(valid_score, 2)) 
          + ' '*(1 + len('Valid') - len(str(round(valid_score, 2)))) + '| '
          + str(round(test_score, 2))
          + ' '*(1 + len('Test') - len(str(round(test_score, 2)))) + '| '
          + str(gs.best_params_['n_estimators'])
          + ' '*(1 + len('n_estimators') - len(str(gs.best_params_['n_estimators']))) + '| '
          + str(gs.best_params_['max_depth']))
    print('-'*(longest_name + header_tail_len + 2))
    best_regressors.append(gs)
    
### END Solution
```

MSE on the training set is the smallest as the model is fitted to this data so the model found patters in this data and can predict well on it. 

In order to prevent overfitting on the training set I performed 5-fold cross validation. This means that model was learned on 80% of training data and validated on the remaining 20% of training set 5 times (each time validation and training subsets were different -- according to cross validation method). Validation score in the table above is the mean value of 5 scores, which were computed during cross validation. This validation score is much higher than the training score because the regressor was taught on 80% of training set and so fitted to this 80% (found patterns in this 80% of training data). So the remaining 20% was new for predictor and the score on it showed the generalizing ability of the trained regressor (it means that the model is able to predict not only on the known training subset). The optimal hyperparameters were chosen according to the best corresponding validation score.

After the optimal hyperparameters were found and the model was trained and validated, the score on the test data was computed. The size of test set is 25% of the whole dataset. The score on test set shows the ability of regressor to predict on the totaly new data. The validation and train scores are approximately the same.

As it can be seen the given dataset is imbalanced. So I chose balanced accuracy as the metric to optimize during CV as it is defined as the average of recall obtained on each class and so deal with imbalanced datasets.

```python
### BEGIN Solution
cv_scores_dict = {
    LogisticRegression.__name__: [0., 0, 0,], 
    RandomForestClassifier.__name__: [0., 0, 0,],
    KNeighborsClassifier.__name__: [0., 0, 0,], 
    DecisionTreeClassifier.__name__: [0., 0, 0,]
}
runtime_dict = {
    LogisticRegression.__name__: [0., 0, 0,], 
    RandomForestClassifier.__name__: [0., 0, 0,],
    KNeighborsClassifier.__name__: [0., 0, 0,], 
    DecisionTreeClassifier.__name__: [0., 0, 0,]
}
index = ['Multiclass', 'OneVsRest', 'OneVsOne']

cv_scores_df = pd.DataFrame(cv_scores_dict, index=index)
runtime_df = pd.DataFrame(runtime_dict, index=index)

clfs = [
    DecisionTreeClassifier(max_depth=7, random_state=RANDOM_STATE),
    RandomForestClassifier(n_estimators=100, max_depth=3, random_state=RANDOM_STATE),
    KNeighborsClassifier(n_neighbors=5),
    LogisticRegression(random_state=RANDOM_STATE)
]

X = StandardScaler().fit_transform(X, y)
for clf in clfs:
    
    # Multiclass labeling approach
    time_start = time.time()
    clf.fit(X, y)
    multiclass_score = cross_val_score(clf, X=X, y=y, cv=5, scoring='balanced_accuracy').mean()
    result_time = round(time.time() - time_start, 7)
    cv_scores_df[type(clf).__name__][index[0]] = multiclass_score
    runtime_df[type(clf).__name__][index[0]] = result_time
    
    # OneVsRest approach
    one_vs_rest = OneVsRestClassifier(clf)
    time_start = time.time()
    one_vs_rest.fit(X, y)
    one_vs_rest_valid_score = cross_val_score(one_vs_rest, X=X, y=y, cv=5, scoring='balanced_accuracy').mean()
    result_time = round(time.time() - time_start, 7)
    cv_scores_df[type(clf).__name__][index[1]] = one_vs_rest_valid_score
    runtime_df[type(clf).__name__][index[1]] = result_time
    
    # OneVsOne approach
    one_vs_one = OneVsOneClassifier(clf)
    time_start = time.time()
    one_vs_one.fit(X, y)
    one_vs_one_valid_score = cross_val_score(one_vs_one, X=X, y=y, cv=5, scoring='balanced_accuracy').mean()
    result_time = round(time.time() - time_start, 7)
    cv_scores_df[type(clf).__name__][index[2]] = one_vs_one_valid_score
    runtime_df[type(clf).__name__][index[2]] = result_time
    
    
print ("Cross Validation Scores")
display(HTML(cv_scores_df.to_html()))
print ("Training Duration (CV=5)")
display(HTML(runtime_df.to_html()))
### END Solution
```

For Logistic Regression OneVsOne approach gave a little more accurate result.

For Random Forest both OneVsRest and OneVsOne approaches gave much better result than multiclass approach.

The result of K-Neighbors Classifier was not affected by applying OneVsRest and OneVsOne approaches. All three approaches gave approximately the same result (OneVsOne was however the worst).

OneVsOne approach applied to Decision Tree Classifier gave signifficant better result than two other approaches.

However if to speak about increase in quality without signifficant increase in computation time only Decision Tree Classifier with OneVsOne performed well: the computation time is of the same order as for multiclass approach while the balanced accuracy increased on 0.15. Logistic Regression with OneVsOne also gave better score with the same order of computation time (although increase in score is not so signifficant).
